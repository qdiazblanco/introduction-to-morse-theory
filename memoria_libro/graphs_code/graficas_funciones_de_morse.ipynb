```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import LightSource


mpl.rcParams.update({
    "font.family": "serif",
    "mathtext.fontset": "cm",
})

# Dominio
x = np.linspace(-1.6, 1.6, 350)
y = np.linspace(-1.6, 1.6, 350)
X, Y = np.meshgrid(x, y)
Z = (X**2 - 1)**2 + (Y**2 - 1)**2

# 🔑 Z comprimido SOLO para sombreado
Z_shade = np.log1p(Z)

fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111, projection='3d')

# Luz suave
ls = LightSource(azdeg=45, altdeg=35) #45,35 original

rgb = ls.shade(
    Z_shade,
    cmap=plt.cm.Greys,
    vert_exag=0.25, #0.25
    blend_mode='overlay'
)

ax.plot_surface(
    X, Y, Z,
    facecolors=rgb,
    linewidth=0,
    antialiased=True,
    shade=False
)

# Puntos críticos
crit = [(-1,-1), (-1,0), (-1,1),
        (0,-1),  (0,0),  (0,1),
        (1,-1),  (1,0),  (1,1)]

cx = [p[0] for p in crit]
cy = [p[1] for p in crit]
cz = [(x**2 - 1)**2 + (y**2 - 1)**2 for x,y in crit]

ax.scatter(cx, cy, cz, color="black", s=22, depthshade=False)

# Limpieza
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.grid(False)

ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$f(x,y)$")

ax.view_init(elev=30, azim=48)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import LightSource, LinearSegmentedColormap

# Estilo tipográfico
mpl.rcParams.update({
    "font.family": "serif",
    "mathtext.fontset": "cm",
})

# Dominio
x = np.linspace(-1.6, 1.6, 400)
y = np.linspace(-1.6, 1.6, 400)
X, Y = np.meshgrid(x, y)
Z = X**3 - 3*X*Y**2

# Z comprimido SOLO para sombreado
Z_shade = np.log1p(np.abs(Z)) * np.sign(Z)

# Colormap gris mate (evita blancos puros)
greys_matte = LinearSegmentedColormap.from_list(
    "greys_matte",
    plt.cm.Greys(np.linspace(0.15, 0.9, 256))
)

fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111, projection='3d')

# Luz suave (igual que antes)
ls = LightSource(azdeg=45, altdeg=2)

rgb = ls.shade(
    Z_shade,
    cmap=greys_matte,
    vert_exag=0.000008,
    blend_mode='overlay'
)

ax.plot_surface(
    X, Y, Z,
    facecolors=rgb,
    linewidth=0,
    antialiased=True,
    shade=False
)

# Punto crítico (origen)
ax.scatter(
    [0], [0], [0],
    color="black",
    s=28,
    depthshade=False
)

# Limpieza total (estilo libro)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.grid(False)

ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$f(x,y)$")

# Perspectiva que deja ver las 3 direcciones
ax.view_init(elev=15, azim=-60)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import LightSource, LinearSegmentedColormap

# Estilo tipográfico (igual que antes)
mpl.rcParams.update({
    "font.family": "serif",
    "mathtext.fontset": "cm",
})

# Dominio
x = np.linspace(-1.6, 1.6, 400)
y = np.linspace(-1.6, 1.6, 400)
X, Y = np.meshgrid(x, y)
Z = (X**2) * (Y**2)

# Z comprimido SOLO para sombreado (evita blanqueo en z=0)
Z_shade = np.log1p(Z)

# Colormap gris mate
greys_matte = LinearSegmentedColormap.from_list(
    "greys_matte",
    plt.cm.Greys(np.linspace(0.15, 0.9, 256))
)

fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111, projection='3d')

# Luz suave (misma que en las otras figuras)
ls = LightSource(azdeg=-50, altdeg=10)

rgb = ls.shade(
    Z_shade,
    cmap=greys_matte,
    vert_exag=0.00000000001,
    blend_mode='overlay'
)

# Superficie
ax.plot_surface(
    X, Y, Z,
    facecolors=rgb,
    linewidth=0,
    antialiased=True,
    shade=False
)

# Conjunto crítico: ejes x=0 e y=0
t = np.linspace(-1.6, 1.6, 200)

# x = 0
ax.plot(
    np.zeros_like(t), t, np.zeros_like(t),
    color="black", linewidth=1.2
)

# y = 0
ax.plot(
    t, np.zeros_like(t), np.zeros_like(t),
    color="black", linewidth=1.2
)

# Limpieza (estilo libro)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.grid(False)

ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$f(x,y)$")

# Perspectiva buena para ver el cruce
ax.view_init(elev=22, azim=-65)

plt.tight_layout()
#ax.set_box_aspect((1, 1, 0.8))
plt.show()
```
